```python
import os
import csv

def create_test_csv(output_file, image_folder):
    with open(output_file, 'w', newline='') as file:
        csv_writer = csv.writer(file)
        csv_writer.writerow(['Image_File'])  # Writing the header

        # Iterate through the images in the folder
        for filename in os.listdir(image_folder):
            if filename.endswith(('.jpg', '.jpeg', '.png')):
                csv_writer.writerow([filename])

if __name__ == "__main__":
    output_csv_file = '/content/drive/MyDrive/Animal_Prediction/Test/test2.csv'
    testing_image_folder = '/content/drive/MyDrive/Animal_Prediction/Test/images2'

    create_test_csv(output_csv_file, testing_image_folder)
```
